```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Pandas/Pandas_projetos_estudos/dados_residencial.csv', sep = ';')
```

# Paraq criar uma nova coluna usando dados já existentes na DataFrame, usamos a DataFrame 'dados' e entre [ ] damos o nome da nova variável e igualamos com as variáveis que vão formar essa nova coluna.


```python
# Valor bruto é a soma da coluna 'Valor', 'Condominio' e 'IPTU'.
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
# Valor por metro quadrado é a divisão da coluna 'Valor' pola coluna 'Area'.
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

# Para usar apenas duas casas decimais, usamos o comando '.round(2).

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados.head(10)
```

```python
# Valor Bruto por metro quadrado.
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

# Para criar uma coluna separando o que é cas do que é apartamento, criamos uma variável 'casa' para definir o que será considerado casa.
# Logo criamos a coluna 'Tipo de Imovel' onde usamos a coluna 'Tipo' com o método'.apply( )' e com a função 'lambda' onde 'x' será 'Casa' se 'x' pertencer a variável 'casa' se não, será 'Apartamento'.

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo de Imovel'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

# excluindo colunas(Variáveis da DataFrame).
# Vamor criar uma DataFrame para não mexer na Tabela original.




```python
dados_aux = pd.DataFrame(dados[['Tipo de Imovel', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

# A primeiro forma de excluir uma coluna é usando 'del', o nome da DataFrame, no caso 'dados_aux' e entre [' '] o nome da coluna.

```python
del dados_aux['Valor Bruto']
dados_aux.head(10)
```

# A outra forma é usando o método '.pop'. Usamos o nome da DataFrame seguido do método '.pop' e entre (' ') o nome da coluna a ser excluida.
# Assim que for executado, o retorno será dos dados que foram exçluídos.

```python
dados_aux.pop('Valor Bruto m2')
dados_aux.head(10)
```

# Também pode ser usado o comando '.drop( )'.
# No camando '.drop( )' temos que passar qual eixo(axis = ) queremos excluir, sendo 0 para linhas e 1 para colunas e implementamos com o método 'inplace = True' para modificar definitivamente a DataFrame.
# Agora vamos usar a DataFrame principal e substituir no nosso banco.

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis= 1, inplace= True)
```
