# Execício

## A variável números tem i para i dentro de range(11), ou seja i tem valores de 0 10.
## letras. A função chr( ) transforma números no seu caracter correspondente. Logo i corresponde ao caracter 65(A) dentro do range(11), de 0 a 11 sendo A igual a 0 e assim por diante.
## Então df é formado por data, os dados da tabela, igual a 0 até 10, a coluno se chama N e o índice vai de A a K.

```python
numeros = [i for i in range(11)]
letras = [chr(i + 65) for i in range(11)]
nome_coluna = ['N']

df = pd.DataFrame(data = numeros, index = letras, columns = nome_coluna)
```

```python
df
```

## A variável 'selecao' usa a coluna 'N' da DataFrame df com a função isin( ) que seleciona 'i' para range(11) se 'i' for par e 0, ou seja, todo o resto(%) de 'i' dividido por 2 tem que ser igual(==) a 0 para que a condição seja válida. 

```python
selecao = df['N'].isin([i for i in range(11) if i % 2 == 0])
df = df[selecao]
df
```
