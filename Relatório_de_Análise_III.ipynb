# Listar a coluna 'tipo' com a função '.drop_duplicates' para que não haja duplicatas.

```python
list(dados['Tipo'].drop_duplicates())
```

# Vou copiar essa lista e tirar todos os tipos de imóveis que são residenciais.

```python
residencial =['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
dados.head(10)
```

# Usar a função '.isin( )' para verificar com 'True' e 'False' os Tipos de imóveis classificados como residencial na DataFrame 'dados'.
# Para isso, é chamado do tabela 'dados' a coluna 'Tipo' e comparado, usando a função '.isin( )' com a varíavel 'residencial', trezendo o resultado 'True' e 'False'.

```python
dados['Tipo'].isin(residencial).head(10)
```

# Criar uma variável desse método.

```python
selecao = dados['Tipo'].isin(residencial)
```

```python
selecao
```

# Criar um novo DataFrame usando apenas o que for 'True' da variável 'selecao'.

```python
dados_residencial = dados[selecao]
```

```python
dados_residencial
```

# Verificando se os dados da nova DataFrame correspondem com o desejado.


```python
list(dados_residencial['Tipo'].drop_duplicates())
```
