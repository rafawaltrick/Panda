```python
america_do_sul = ['Venezuela', 'Brasil', 'Peru', 'Bolivia', 'Chile', 'Argentina', 'Ecuador', 'Paraguay', 'Colombia', 'Uruguay', 'Guyana', 'Suriname']
```

```python
america_do_sul
```

```python
tabela_inpe['pais'].isin(america_do_sul)
```

```python
selecao_a_s = tabela_inpe['pais'].isin(america_do_sul)
```

```python
tabela_america_do_sul = tabela_inpe[selecao_a_s].copy()
```

```python
tabela_america_do_sul
```

```python
tabela_america_do_sul['pais'].drop_duplicates()
```

```python
tabela_america_do_sul['frp']
```

```python
tabela_america_do_sul = tabela_america_do_sul.fillna({'frp': 0})
```

```python
tabela_america_do_sul['frp']
```

```python
grupo_pais = tabela_america_do_sul.groupby('pais')
```

```python
grupo_pais
```

```python
grupo_pais['frp']
```

```python
grupo_pais['frp'].describe()
```

```python
grupo_pais['frp'].max().plot.bar(color = 'red')
plt.title('Tabela Valor Máximo "frp"')
plt.xlabel('Países')
plt.ylabel('Valores Max FRP')
```

```python
grupo_pais['risco_fogo'].describe()
```

```python
grupo_pais['risco_fogo'].max().plot.bar(color = 'orange')
plt.title('Tabela Valor Máximo "Risco de Fogo"')
plt.xlabel('Países')
plt.ylabel('Valores Max Risco de Fogo')
plt.grid(True)
```
