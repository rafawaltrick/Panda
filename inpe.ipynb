```python
# A variável "a" é uma lista.
a = ['Zambia', 'Angola', 'Republic of Congo', 'Egypt', 'Mozambique',
       'South Africa', 'Sudan', 'Peru', 'Bolivia', 'Chad', 'Chile',
       'Argentina', 'Mexico', 'Russia', 'Turkey', 'Bulgaria',
       'Bosnia and Herzegovina']
```

```python
# Edição do primeiro índice da lista "a".
a[0] = 'jacarei'
```

```python
# Retorno da lista "a" com a alteração.
a
```

```python
# "b" é uma "tuple" de "a". Logo "b" não pode ser alterado, editado.
b = tuple(a)
```

```python
type(b)
```

```python
# Por ser uma 'tuple', não pode ser alterado.  (TypeError: 'tuple' object does not support item assignment) 
b[0] = 'Zambia'
```

```python
b
```
