<h1>Criada uma Tabela com  nome "venda" com 4 colunas e cada coluna com uma lista representada por [ ]</h1><rn> <h1>Criada a variavel "tabela_venda" e vinculada a biblioteca Pandas.</h1>


```python


venda = {'data': ['15/02/2022', '17/03/2022'],
         'valor': [350, 546],
         'quantidade': [2, 1],
         'produto': ['memória', 'HD'],
         
        }


tabela_vendas = pd.DataFrame(venda)


display(tabela_vendas)
```
