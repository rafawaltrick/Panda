```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
dados_12_2022 = 'https://queimadas.dgi.inpe.br/home/downloadfile?path=%2Fapp%2Fapi%2Fdata%2Fdados_abertos%2Ffocos%2FMensal%2Ffocos_abertos_mensal_202212.csv'
```

```python
dados_11_2022 = 'https://queimadas.dgi.inpe.br/home/downloadfile?path=%2Fapp%2Fapi%2Fdata%2Fdados_abertos%2Ffocos%2FMensal%2Ffocos_abertos_mensal_202211.csv'
```

```python
dados_11 =pd.read_csv(dados_11_2022)
```

```python
dados_12 = pd.read_csv(dados_12_2022)
```

```python
america_do_sul = ['Venezuela', 'Brasil', 'Peru', 'Bolivia', 'Chile', 'Argentina', 'Ecuador', 'Paraguay', 'Colombia', 'Uruguay', 'Guyana', 'Suriname']
```

```python
paises_america_sul_11 = dados_11['pais'].isin(america_do_sul)
```

```python
paises_america_sul_12 = dados_12['pais'].isin(america_do_sul)
```

```python
tabela_paises_america_sul_11 = dados_11[paises_america_sul_11].copy()
```

```python
tabela_paises_america_sul_12 = dados_12[paises_america_sul_12].copy()
```

```python
tabela_paises_america_sul_11
```

```python
tabela_paises_america_sul_11.pais.unique()
```

```python
pd.merge(tabela_paises_america_sul_11, tabela_paises_america_sul_12, how = 'outer', on = 'precipitacao')
```

```python
grupo_america_sul_11 = tabela_paises_america_sul_11.groupby('pais')
```

```python
grupo_america_sul_11
```

```python
grupo_america_sul_11['frp', 'precipitacao'].max().plot.bar()
```

```python
grupo_biomas_11 = tabela_paises_america_sul_11.groupby('bioma')
```

```python
grupo_biomas_11['frp', 'precipitacao'].max().plot.bar()
```

```python
grupo_municipios_11 = tabela_paises_america_sul_11.groupby('municipio')
```

```python
tabela_paises_america_sul_11.municipio.unique().shape
```
