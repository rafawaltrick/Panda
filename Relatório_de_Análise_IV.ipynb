```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Pandas/Pandas_projetos_estudos/dados_residencial.csv', sep=';')
```

```python
dados.head(10)
```

# Na variável 'selecao' o retorno é apenas a coluna 'Tipo', sendo True para os tipos Apartamento e False para os outros tipos.

```python
# Selecione somente os imóveis classificados com tipo 'Apartamento'.
selecao = dados['Tipo'] == 'Apartamento'
selecao
```

# Na variável 'n1', o retorno são apenas os Apartamento(True) selecionados na variável 'selecao.

```python
n1 = dados[selecao]
n1
```

# Adicionado a função '.shape[0]' para saber o tamanho, a quantidade de linhas. Se usasse '.shape[1]', o retorno seria o número de colunas.

```python
n1 = dados[selecao].shape[0]
n1
```

# Com a variável 'selecao2', quero selecionar mais de 1 item da coluna 'Tipo'. Sendo assim, tenho que separar por '|' que significa 'ou' e adicionar ( ) entre os termos.

```python
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
selecao2 = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
selecao2
```

#  Variável 'n2' para criar a DataFrame com os itens selecionados.

```python
n2 = dados[selecao2]
n2
```

## '.shape[0]' para ver o tamanho, a quantidade de linhas da DataFrame.

```python
n2 = dados[selecao2].shape[0]
n2
```

# Na 'selecao3' tenho que selecionar entre números. Usando ( ) para cada condição ligadas pelo termo '&', simbolizando que ambos tem que ser cumpridos para a condição valer.

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
selecao3 = (dados['Area'] >= 60) & (dados['Area'] <= 100) 
selecao3
```

```python
n3 =dados[selecao3]
n3
```

```python
n3 = dados[selecao3].shape[0]
n3
```

# Nessa seleção, foram usadas 2 condições de colunas diferentes. Mas o procedimento é mesmo.

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.
selecao4 = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000.0)
selecao4
```

# Com a variável n4, será retornado a DataFrame com as condições pedidas e o número de linhas usando a função '.shape[0]'.

```python
n4 = dados[selecao4]
n4
```

```python
n4 = dados[selecao4].shape[0]
n4
```

# Criando linhas de texto para descrever e usar os dados criados.

```python
print("Nº de imóveis classificados com tipo 'Apartamento' -> {}".format(n1))
print("Nº de imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila' -> {}".format(n2))
print("Nº de imóveis com área entre 60 e 100 metros quadrados, incluindo os limites -> {}".format(n3))
print("Nº de  imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00 -> {}".format(n4))
```
