```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Pandas/Pandas_projetos_estudos/dados_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

# Para identificarmos os dados faltantes, podemos usar '.isnull( )'. Com ele ira ser retornado 'False' para os espaços ocupados e 'True' para os espaços vazios.

```python
dados.isnull()
```

# Temos também o '.notnull( )'. Que exatamente o oposto do comando anterior.




```python
dados.notnull()
```

# Como a função '.info', podemos verificar que as 6 primeiras variáveis, não tem espaços vazios. Porem as variáveis 'Valor', 'Condomonio' e 'IPTU' possuem espaçõs vazios.
# Na variável valor é mais fácil de visualizar esses espaços, por ser um número pequeno de espaços vazios, são 9 espaços vazios.

```python
dados.info()
```

# Aqui foi selecionado da coluna 'Valor' os valores nulos com 'True' e os não nulos com 'False'.

```python
dados['Valor'].isnull()
```

# Agora foi retornado dentro da DataFrame, apenas os valores nulos da coluna 'Valor'.

```python
dados[dados['Valor'].isnull()]
```

# Criando a variável 'A' para retornar o número de dados da DataFrame.
# Usando o método '.dropna( ), eu elimino os dados nulos e com o argumento 'subset = (' ')' que cria uma lista com a ou as variáveis que eu desejo eliminar os valores nulos. Somado ao 'inplace = True' para fixar a nova modificação na DataFrame.
# E criamos a variável 'B' para mostrar os novos valores da DataFrame.
# Sendo assim, 'A' - 'B' terá uqe retornar o valor de elementos que foi retirado da DataFrame. 

```python
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

# Chamarmos, de novo, os valores nulos da variável 'Valor', o retorno será uma DataFrame vazia.

```python
dados[dados['Valor'].isnull()]
```

# Assumindo que todo o apartamento tem condomínio.
# Primeiro vamos visualizar a coluna 'Condominio' na DataFrame dados que tenha valores nulos.
# Somando o método '.shape[0], temos o número de condominio nulos.

```python
dados[dados['Condominio'].isnull()]
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

# Criando uma variável para selecionar da DataFrame 'dados' a coluna 'Tipo' que seja igual a apartamento e selecionar da coluna 'Condominio' os dados que sejam nulos.

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

# Como já feito anteriormente, foi criada uma váriavel para visualizar os dados antigos.
# Depois foi recondicionado a variável 'dados' como o valor ad variável 'selecao' que mostra justamente os apartamentos com condomínio nulo e por isso é colocado o símbolo '~' para inverter o valor da variável 'selecao' e retornar os apartamentos com valor não nulos.
# Foi criada um avariável para retornar o nov valor da variável 'dados'.
# E por último a subtração das variáveis para mostrar que foram retirados os valores nulos da DataFrame.

```python
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

# Para manter os valores nulos da coluna 'Condominio' mas com valor 0, podemos usar o camando '.fillna( )' atribuindo o valor 0 dentro dos ( ).

```python
dados.fillna(0, inplace = True)
```

# Ou podemes usar um dicionario para adicionar nas colunas onde tem espaços vazios o valor desejado.

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
```

# Dessa forma podemos ver que não existem mais valores nulos na coluna "Condominio' e nem na coluna "IPTU'.

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[dados['IPTU'].isnull()].shape[0]
```

# Vendo as informaçães da DataFrame percebemos que não existem mais valores nulos.

```python
dados.info()
```
